```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=sns.load_dataset('titanic')
ds
```

```python
print(ds.head())
print(ds.tail())
print(ds.info())
print(ds.describe(include="all"))
print(ds.shape)
print(ds.size)
```

```python
print(ds.isnull())
print(ds.isnull().sum())
```

```python
print(sns.displot(ds['fare']))
print(sns.displot(ds['fare'],kde=False,bins=10))
```

```python
sns.histplot(ds['fare'], kde=False, bins=10)
plt.show()
```

```python
sns.jointplot(x='age', y='fare', data=ds)
```

```python
sns.jointplot(x='age', y='fare', data=ds, kind='hex')
```

```python
sns.pairplot(ds, hue='sex')
plt.show()
```

```python
sns.rugplot(ds['fare'])
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=ds,estimator=np.std)
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=ds, hue="survived")
plt.show()
```

```python
sns.stripplot(x='sex', y='age', data=ds, jitter=True, hue='survived')
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=ds, hue='survived', split=True)
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=ds)
plt.show()
```

```python
sns.stripplot(x='sex', y='age', data=ds, hue='survived',jitter=True,dodge=True)
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', data=ds, hue='survived', dodge=True)
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=ds)
sns.swarmplot(x='sex', y='age', data=ds, color='black')
plt.show()
```
