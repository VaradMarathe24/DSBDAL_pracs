```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
print(titanic.head())
print(titanic.tail())
print(titanic.info())
print(titanic.describe())
print(titanic.isnull())
```

```python
print(titanic.isnull().any())
print("---------------------------")
print(titanic.isnull().sum())
```

```python
sns.boxplot(x="sex",y="age",data=titanic)
```

```python
sns.boxplot(x="sex",y="age",hue="survived",data=titanic)
```

```python
sns.distplot(titanic[titanic["survived"]==0]["age"],hist=False,color='blue')
sns.distplot(titanic[titanic["survived"]==1]["age"],hist=False,color='red')
```

```python
sns.countplot(x="survived",data=titanic)
plt.show()
```

```python
titanic['sex'].value_counts().plot(kind="pie",autopct="%.2f")
```

```python
sns.histplot(['age'],bins=5)
```

```python
sns.barplot(x='pclass',y='fare',hue='sex',data=titanic)
```

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['pclass'],titanic['survived']))
```
